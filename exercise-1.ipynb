```python
# Initializing parameter vector theta
import numpy as np
theta = np.zeros((2,1))
print(theta)

theta.shape
```

#### Dataframe.values --> returns a numpy representation of the dataframe 
### Only the values in the DataFrame will be returned, the axes labels will be removed.


```python
X = X.values
y = y.values
X.shape
```

```python
m = len(X)
# Our cost function
def compute_cost(X,y,theta):
    h = np.dot(X, theta) # (97,1)
    error = np.power((h - y), 2)
    J = (1/(2*m)) * np.sum(error)
    return J
```

```python
Cost = compute_cost(X,y,theta)
print('Cost = ' + str(Cost))
```

```python
# Gradient Descent
m = len(X)
def Grad_descent(X, y, theta, alpha, iters):
    
    J_history = np.zeros((iters,1))
    
    for i in range(iters):
            h = np.dot(X, theta)
            error = h - y
            delta = np.dot(X.T, error)
            theta = theta - ((alpha / m) * delta)
            
            # Save the cost J in every iteration
            J_history[i] = compute_cost(X,y,theta)
            
    return theta, J_history
```

```python
# Initialize variables for learning rate and iterations
theta = np.zeros((2,1))
alpha = 0.01
iters = 1500


theta, J_history = Grad_descent(X, y, theta, alpha, iters)
print('theta = ' + str(theta))
```

```python
# The theta vector describes the optimal linear model foe our datset. Let's check our cost/error to check how good our
# evaluation model is
compute_cost(X,y,theta)
```

That's a lot better than 32. But it's not a intuitive way to look at it.

```python
# Predict Profit for the cityu with populkation of 35,000 and 7,000
# h = X * theta
profit1 = np.dot([1, 3.5], theta) * 10000
profit2 = np.dot([1, 7], theta) * 10000

print(profit1)
print(profit2)
```
